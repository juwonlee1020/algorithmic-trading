## Retuns
In calculating the daily returns, we can make use of the pct_change function in the Pandas package to get the change in the value of a stock over a single day of trading. This value will be the percent change in the value of a stock over a single day of trading without taking into account dividents.

```python
import numpy as np
```

```python
daily_close = aapl[['Adj. Close']]
```

```python
daily_pct_change = daily_close.pct_change()
```

```python
daily_pct_change.fillna(0, inplace=True)
```

```python
print(daily_pct_change)
```

```python
daily_log_returns = np.log(daily_close.pct_change() + 1)
daily_log_returns
```

```python
#Resampling to business months
monthly = aapl.resample('BM')
monthly.head()
```

```python
monthly.apply(lambda x:x[-1])
monthly.pct_change()
```

```python
quarter = aapl.resample("4M").mean()
```

```python
quarter.pct_change()
```

```python
daily_pct_change = daily_close / daily_close.shift(1) - 1
daily_pct_change
```

```python
daily_log_returns_shift = np.log(daily_close / daily_close.shift(1))
daily_log_returns_shift
```

```python
#Plotting the distribution of daily_pct_change
daily_pct_change.hist(bins=50)
plt.show()
```
